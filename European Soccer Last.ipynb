```python
home_GD = match.home_team_goal - match.away_team_goal
match['home_GD'] = home_GD
away_GD = match.away_team_goal - match.home_team_goal
match['away_GD'] = away_GD
```

```python
def check_value3(cell_value = 0):
    if cell_value > 0:
        return 'w'
    elif cell_value == 0:
        return 'd'
    else:
        return 'l'
match['home_stat'] = match['home_GD'].apply(check_value3)
```

```python
match['away_stat'] = match['away_GD'].apply(check_value3)
```

```python
def check_value2(cell_value = 'w'):
    if cell_value =='w':
        return 3
    elif cell_value =='d':
        return 1
    else:
        return 0
match['home_points'] = match['home_stat'].apply(check_value2)
```

```python
match['away_points'] = match['away_stat'].apply(check_value2)
```

```python
def check_value(cell_value = 'w'):
    if cell_value == 'w':
        return 1
    else:
        return 0
match['home_win'] = match['home_stat'].apply(check_value)
```

```python
match['away_win'] = match['away_stat'].apply(check_value)
```

```python
#Merge Match with Country
MatchMerged=match.merge(country,left_on='country_id',right_on='id',how='left',suffixes=('match','country'))
MatchMerged.rename(columns={'name':'country'},inplace=True)

#Merge with League info
MatchMerged=MatchMerged.merge(league,left_on='league_id',right_on='country_id',how='left',suffixes=('match','league'))
MatchMerged.rename(columns={'id':'id_league','name':'league'},inplace=True)

#Merge with Team Info
MatchMerged=MatchMerged.merge(team[['team_api_id','team_long_name']],left_on='home_team_api_id',right_on='team_api_id',how='left',suffixes=('match','team'))
MatchMerged.rename(columns={'team_api_id':'team_api_id_home','team_long_name':'home_team'},inplace=True)
MatchMerged=MatchMerged.merge(team[['team_api_id','team_long_name']],left_on='away_team_api_id',right_on='team_api_id',how='left',suffixes=('match','away'))
MatchMerged.rename(columns={'team_api_id':'team_api_id_awai','team_long_name':'away_team'},inplace=True)
```

```python
#Split the match dataframe to get home team & away team statistics
home_team=MatchMerged[['season','league_id','league_name','home_team_api_id','home_team','away_team','home_team_goal','away_team_goal']]
away_team=MatchMerged[['season','league_id','league_name','away_team_api_id','away_team','home_team','away_team_goal','home_team_goal']]

home_team.rename(columns={'home_team_api_id':'team_api_id',
                 'home_team':'team',
                 'away_team':'opp_team',
                 'home_team_goal':'team_goal',
                 'away_team_goal':'opp_team_goal'},inplace=True)
away_team.rename(columns={'away_team_api_id':'team_api_id',
                 'away_team':'team',
                 'home_team':'opp_team',
                 'away_team_goal':'team_goal',
                 'home_team_goal':'opp_team_goal'},inplace=True)

TeamPoints=pd.concat([home_team,away_team],axis=0,ignore_index=True)

#Create 'Goal Difference' column
TeamPoints['goal_diff']=TeamPoints['team_goal']-TeamPoints['opp_team_goal']

#Create 'team_points' column for each team with following points system, (3: Winning Team, 1: Draw)
TeamPoints.loc[TeamPoints['team_goal']>TeamPoints['opp_team_goal'],'team_points']=3
TeamPoints.loc[TeamPoints['team_goal']==TeamPoints['opp_team_goal'],'team_points']=1

#Create 'Wins', 'Draws' & 'Losses' column 
TeamPoints.loc[TeamPoints['team_goal']>TeamPoints['opp_team_goal'],'wins']=1
TeamPoints.loc[TeamPoints['team_goal']<TeamPoints['opp_team_goal'],'losses']=1
TeamPoints.loc[TeamPoints['team_goal']==TeamPoints['opp_team_goal'],'draw']=1

#Capture matches played
TeamPoints['matches_played']=1

#Rearranging the tables to match actual score tables
TeamPoints.reindex(columns=['season', 'league_name', 'team','team_api_id', 'opp_team', 'matches_played',
'wins','draw','losses','team_goal','opp_team_goal','goal_diff','team_points'],copy=False)

#Generate Score table, by grouping the dataframe on 'league','season','team','team_api_id'. Also, let's sort it based on the following parameters in the below mentioned order
ScoreTable=TeamPoints.groupby(['league_id','league_name','season','team','team_api_id']).sum().sort_values(['league_id','league_name','season','team_points','goal_diff','team_goal'],ascending=[True,True,False,False,False,False]).reset_index()
```

```python
#Group the Match statistics on the basis of the Teams, and create new columns to calculate winning & losing proportion
Consolidated=ScoreTable.groupby(['team']).sum()
Consolidated['win_prop']=round(Consolidated['wins']/Consolidated['matches_played']*100,2)
Consolidated['loss_prop']=round(Consolidated['losses']/Consolidated['matches_played']*100,2)

#Sort based on columns win_prop & loss_prop to find the Most Winning & Losing teams respectively
MostWinningTeam=Consolidated.sort_values('win_prop',ascending=True).tail(10)['win_prop']
MostLosingTeam=Consolidated.sort_values('loss_prop',ascending=True).tail(10)['loss_prop']

plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
MostWinningTeam.plot.barh(color=sns.color_palette('Greens', 10),edgecolor='k')
plt.title('Most Winning Team in History',fontsize=20)
plt.xlim(0,100)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('% of Wins',fontsize=15)
plt.ylabel('Teams',fontsize=15)
for i,j in enumerate(MostWinningTeam.index):
    plt.text(MostWinningTeam.loc[j]+1.5,i,str(MostWinningTeam.loc[j])+'%',style='italic',fontsize=12)    

plt.subplot(1,2,2)
MostLosingTeam.plot.barh(color=sns.color_palette('Reds', 10),edgecolor='k')
plt.title('Most Losing Team in History',fontsize=20)
plt.xlim(0,100)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('% of Losses',fontsize=15)
plt.ylabel('')
for i,j in enumerate(MostLosingTeam.index):
    plt.text(MostLosingTeam.loc[j]+1.5,i,str(MostLosingTeam.loc[j])+'%',style='italic',fontsize=12)

plt.subplots_adjust(wspace=0.5)
```
